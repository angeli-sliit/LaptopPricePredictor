```python
import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/angeli-sliit/LaptopPricePredictor/main/dataset/laptop_price.csv'
data = pd.read_csv(url, encoding='latin-1')
```

```python
data.shape
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['Ram'] = data['Ram'].astype(str).str.replace('GB', '').astype('int32')
data['Weight'] = data['Weight'].astype(str).str.replace('kg', '').astype('float32')
```

```python
data.corr(numeric_only=True)['Price_euros']
```

```python
(data['Company'].value_counts())
```

```python
def add_Company(input):
    if input == 'Samsung' or input == 'Mediacom' or input== 'Razer' or input== 'Microsoft' or input== 'Vero' or input== 'Xiaomi' or input== 'Chuwi' or input== 'Fujitsu' or input== 'Google' or input== 'LG' or input== 'Huawei':
        return 'Other'
    else:
        return input

data['Company'] = data['Company'].apply(add_Company)
    
```

```python
(data['Company'].value_counts())
```

```python
data.head()
```

```python
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['IPS'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data['cpu_name'] = data['Cpu'].apply(lambda x:' '.join(x.split()[0:3]))
```

```python
def set_processor(name):
    if name == 'Intel Core i7'  or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0]=='AMD':
            return 'AMD'
        else:
            return 'Others'

data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x:' '.join(x.split()[0:1]))
```

```python
data = data[data['gpu_name'] != 'ARM']
```

```python
def set_os(input):
    if input == 'Windows 10' or input == 'Windows 7' or input == 'Windows 10 S':
        return 'Windows'
    elif input == 'macOS' or input == 'Mac OS X':
        return 'Mac'
    elif input == 'Linux':
        return input
    else:
        return 'Others'

data.loc[:, 'OpSys'] = data['OpSys'].apply(set_os)
```

```python
(data['OpSys'].value_counts())
```

```python
data.head(2)
```

```python
data = data.drop(columns=['laptop_ID', 'Inches', 'ScreenResolution', 'Cpu', 'Gpu', 'Product'])
```

```python
data.head(2)
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
x = data.drop('Price_euros', axis=1)
y = data['Price_euros']
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
x_train.shape,  x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train, y_train)
    acc = model.score(x_test, y_test)
    print(str(model)+ '-->'+str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
Lasso = Lasso()
model_acc(Lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_estimators': [10, 50, 100], 
    'criterion': ['squared_error', 'absolute_error', 'poisson']
}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)
grid_fit = grid_obj.fit(x_train, y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test, y_test)
```

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
